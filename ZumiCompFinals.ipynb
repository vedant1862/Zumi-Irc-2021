# Running Functions

```python
#Function used for detecting color diffrence between red and green
def ColorScanner():
    from zumi.util.camera import Camera
    from zumi.util.vision import Vision # Importing libraries used to use camera

    camera = Camera()
    vision = Vision()
    data = None
    try:
        camera.start_camera()
        while data is None: # This part of the code is meant to detect if the zumi's camera detects any green in front of it
            image=camera.capture() # if it does recieve any data indicating it has detected green it will close the camera
                                    # and exit the loop, if not it will continue to play do rae me
            data = vision.find_green_object(image, h_range=15, s_range=40, v_range=15, draw_color=(255, 85, 255),bounding_box=True)
            zumi.play_note(25, 1000)
            zumi.play_note(27, 1000)
            zumi.play_note(29, 1000)
            if data is not None:

                camera.show_image(image)
                camera.clear_output()
    finally:        
        camera.close()
        print("done")
        

def ZumiScanner2(): #This functions main purpose is to see wether or not zumi detects any qr_codes in front of it                
    camera = Camera() # The code runs in a loop till it detects a qr code, and when it does it plays a do ra me sound
    vision = Vision()
    camera.start_camera()
    qr_code = None #This variable is set to none, and is assgined a value after it detects a specific qr_code
    try:
        while qr_code == None:
            time_start = time.time()
            image = camera.capture()

            #find the QR code
            qr_code = vision.find_QR_code(image)

            #print if any QR code is detected
            x = vision.get_QR_message(qr_code)

            #this will also clear any print output in jupyter
            camera.clear_output()
    finally:
        camera.close() #plays do ra me when detects qr_code
        print(x)
        zumi.play_note(25, 1000)
        zumi.play_note(27, 1000)
        zumi.play_note(29, 1000)


    return x




def StopSignScan2(): # function meant to detect whether or not a stop sign is infront of Zumi, while there i
    data = True      # sign in front of zumi it plays do ra me, once removed zumi exits the loop, and moves on 
    try:
        camera.start_camera()
        while data == True:
            image=camera.capture()
            data = vision.detect_stop_sign(image)
            if data is True:
                zumi.stop()
                zumi.play_note(25, 1000)
                zumi.play_note(27, 1000)
                zumi.play_note(29, 1000)
            camera.show_image(image)
            camera.clear_output()
    finally:        
        camera.close()
        print("done")
        
        data = False
        
        
        return data
```

# Part A

```python

def PartA(): #Declaring varaibles that are used throughout the rest of the code
    speed_var = 60
    zumi.reset_gyro()
    lines = 0
    y1 = None
    zumi.reset_gyro()
    #Using Qr_code scanning function and assigning y1 the value/location recived from the Qr_code
    while y1 == None:
        y1 = ZumiScanner2()


        print(y1)
        screen.draw_text_center(y1,font_size=12)

    #use turn right or left degree
    #Depending on the location recived from scanning the Qr_code Zumi will travel to specific locations in Part A
    if y1 == 'factory':
        lines = 4 # Lines is very important for part A, as depending on the location recived from the qr code, zumi 
                #determines the distance needed traveled from amount of lines assigned to it
        zumi.drive_over_markers(road_markers=lines,speed=12,time_out=3)
        zumi.turn(-90, max_speed = 35)
        # After traveling a certain amout of lines zumi turns either left or right (depending on assigned locations) 
        #and stays in location for .5 sec before backing our and continuing on its path till it meets the wall ahead
        #where it turns right
        zumi.forward(speed_var) 
        time.sleep(.5)
        zumi.reverse()
        zumi.turn(0,max_speed = 35)
        while zumi.read_z_angle() > 5:
            zumi.turn_left(1)

        while zumi.read_z_angle() < -5:
            zumi.turn_right(1)


        zumi.forward_avoid_collision(speed_var,5, 0,130,130)

        zumi.turn(-90, max_speed = 35) 
    # at this point Zumi should have turned to face the funnel at the end of street
        #for i in range(10): # can increase range
        #zumi.go_straight(50, -90)

    elif y1 == 'school':
        lines = 8
        zumi.drive_over_markers(road_markers=lines,speed=12,time_out=3)
        zumi.turn(90, max_speed = 35)
        zumi.forward(speed_var) # can adjust later
        time.sleep(.5)
        zumi.reverse()
        zumi.turn(0, max_speed = 35)

        while zumi.read_z_angle() > 5:
            zumi.turn_left(1)

        while zumi.read_z_angle() < -5:
            zumi.turn_right(1)
        zumi.forward_avoid_collision(speed_var,5, 0,130,130)
        zumi.turn(-90, max_speed = 35)
        #for i in range(2): # can increase range
            #zumi.go_straight(50, -90)

    elif y1 == 'office':
        lines = 12
        zumi.drive_over_markers(road_markers=lines,speed=12,time_out=5)
        zumi.turn(-90, max_speed = 35)
        zumi.forward(speed_var) # can adjust later
        time.sleep(.5)
        zumi.reverse()
        zumi.turn(0, max_speed = 35)
        while zumi.read_z_angle() > 5:
            zumi.turn_left(1)

        while zumi.read_z_angle() < -5:
            zumi.turn_right(1)
        zumi.forward_avoid_collision(speed_var,5, 0,130,130)
        zumi.turn(-90, max_speed = 35)
        #for i in range(10): # can increase range
         #   zumi.go_straight(50, -90)

    elif y1 == 'museum':
        lines = 16
        zumi.drive_over_markers(road_markers=lines,speed=12,time_out=6.5)
        zumi.turn(90, max_speed = 35)
        zumi.forward(speed_var) # can adjust later
        time.sleep(.5)
        zumi.reverse()
        zumi.turn(0, max_speed = 35)
        while zumi.read_z_angle() > 5:
            zumi.turn_left(1)

        while zumi.read_z_angle() < -5:
            zumi.turn_right(1)
        #zumi.drive_over_markers(road_markers=2,speed=20,time_out=10)
        zumi.forward_avoid_collision(speed_var,5, 0,130,130)
        zumi.turn(-90, max_speed = 35)



    # zumi goes forward till it meets a funnel section where the funnel_align funtion that runs for 9 seconds at predetermined
    # values at helps it reach the end of the funnel
    for i in range(0, 55):
        zumi.go_straight(45,-90)

    zumi.stop()
    
    
    zumi.funnel_align(speed=2.5, duration=8.5, angle=-90, angle_adj=.5, l_th=90, r_th=90)
   
    # zumi should now at the narrow tip of the funnel facing the stop sign for Section B




```

```python
def PartC():
    
    ColorScanner()




    zumi.forward(20,.5)
    zumi.line_follow_gyro_assist(speed=15, duration=10)
    zumi.turn_left()
    zumi.line_follow_gyro_assist(speed=15, duration=10)
    zumi.turn_right()
    zumi.line_follow_gyro_assist(speed=15, duration=10)
    zumi.turn_right()
    f = ZumiScanner2()

    if f == 'right':

        zumi.line_follow_gyro_assist(speed=15, duration=2.0)
        zumi.turn_right()
        zumi.line_follow_gyro_assist(speed=15, duration=1.0)
        zumi.turn_left()
        zumi.line_follow_gyro_assist(speed=20, duration=7.0)
        zumi.forward(20,1)

    elif f == 'left':
        zumi.line_follow_gyro_assist(speed=15, duration=2.0)
        zumi.turn_left()
        zumi.line_follow_gyro_assist(speed=15, duration=2.0)
        zumi.turn_right()
        zumi.line_follow_gyro_assist(speed=20, duration=6.0)
        zumi.forward(20,1)  


    personality.celebrate()
    zumi.turn_right(360, .1)

    screen.draw_text_center(y1 + "            " + f,font_size=12)


```
